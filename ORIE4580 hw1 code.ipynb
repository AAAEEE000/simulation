```python
from math import factorial as f
from math import pi
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

3(b): plot f as a function of p>0.5 for fixed n=7

```python
def nCr(n,r):
    return f(n) / f(r) / f(n-r)
```

```python
x_1 = np.arange(0.5, 1, 0.01)
y_1 = nCr(7,4)* x_1**4 * (1-x_1)**3  +  nCr(7,5)* x_1**5 * (1-x_1)**2\
      + nCr(7,6)* x_1**6 * (1-x_1)  +  nCr(7,7)* x_1**7
plt.plot(x_1, y_1);
```

The winning probability f increases as p increases, which makes sense. The curve is concave. f finally reaches 1 when p=1.

3(b): plot f for fixed p=0.55 and odd n varying from 1 thru 21

```python
x_2 = np.arange(1, 22, 2) #odd n from 1 thru 21
y_2 = np.zeros(len(x_2))
for p in range(len(x_2)):
    i = (x_2[p]+1)/2
    #print p,i
    for q in range(i, x_2[p]+1):
        y_2[p] += nCr(x_2[p], q) * 0.55**q * (1-0.55)**(x_2[p]-q)
        
#print y_2
plt.plot(x_2, y_2, "--o");
```

The winning probability initially increases very fast with n, but later increases slower and almost linearly. For the stronger team, it is beneficial to play in a longer series.

```python
in_circle_num = np.zeros(1000)
for i in range(1,1001):
    x = np.random.uniform(-1, 1, i)
    y = np.random.uniform(-1, 1, i)
    in_circle_bool = ( (x**2 + y**2)<1 )+0
    in_circle_num[i-1] = np.sum(in_circle_bool)
    
estimated_pi = 4*in_circle_num/range(1,1001)
relative_err = abs(estimated_pi - pi)/pi
inverse_sqrt = 1/np.sqrt(range(1,1001))
plt.plot(range(1,1001), relative_err, range(1,1001), inverse_sqrt);
```

The relative errors is proportional to the inverse sqrt function, and they progresses similarly as sample size gets larger.
